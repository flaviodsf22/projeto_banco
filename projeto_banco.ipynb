```python
#Projeto banco 

def depositar(valor):
    global saldo, extrato
    if valor > 0:
        saldo += valor
        extrato += f"Depósito: R$ {valor:.2f}\n"
    else:
        print("Operação falhou! O valor informado é inválido.")

def sacar(valor):
    global saldo, extrato, numero_saques
    excedeu_saldo = valor > saldo
    excedeu_limite = valor > limite
    excedeu_saques = numero_saques >= LIMITE_SAQUES
    if excedeu_saldo:
        print("Operação falhou! Você não tem saldo suficiente.")
    elif excedeu_limite:
        print("Operação falhou! O valor do saque excede o limite.")
    elif excedeu_saques:
        print("Operação falhou! Número máximo de saques excedido.")
    elif valor > 0:
        saldo -= valor
        extrato += f"Saque: R$ {valor:.2f}\n"
        numero_saques += 1
    else:
        print("Operação falhou! O valor informado é inválido.")

def extrato_bancario():
    global extrato, saldo
    print("\n================ EXTRATO ================")
    print("Não foram realizadas movimentações." if not extrato else extrato)
    print(f"\nSaldo: R$ {saldo:.2f}")
    print("==========================================")

def realizar_operacao(opcao):
    if opcao == "d":
        valor = solicitar_valor("depósito")
        depositar(valor)
    elif opcao == "s":
        valor = solicitar_valor("saque")
        sacar(valor)
    elif opcao == "e":
        extrato_bancario()
    elif opcao == "q":
        return False
    else:
        print("Operação inválida, por favor selecione novamente a operação desejada.")
    return True

def solicitar_valor(tipo_operacao):
    while True:
        try:
            valor = float(input(f"Informe o valor do {tipo_operacao}: "))
            if valor >= 0:
                return valor
            else:
                print("O valor não pode ser negativo.")
        except ValueError:
            print("Por favor, insira um valor numérico válido.")

def menu_principal():
    while True:
        opcao = input(menu)
        if not realizar_operacao(opcao):
            break

saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3
menu = """

[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair

=> """

menu_principal()
```
